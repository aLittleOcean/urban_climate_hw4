```python
import rasterio
import numpy as np
import pandas as pd
```

### По всему региону

```python
src = rasterio.open("lcz.tif")
```

```python
lcz = src.read(1).astype(int)
nodata = src.nodata
mask = (~np.isnan(lcz)) & (lcz != nodata) & (lcz > 0)
lcz = lcz[mask]
```

```python
values, counts = np.unique(lcz, return_counts=True)

df = pd.DataFrame({
    "LCZ": values,
    "pixels": counts})

df["share (%)"] = df["pixels"] / df["pixels"].sum() * 100
```

```python
lcz_names = {
1:"Compact High-Rise",
2:"Compact Mid-Rise",
3:"Compact Low-Rise",
4:"Open High-Rise",
5:"Open Mid-Rise",
6:"Open Low-Rise",
7:"Lightweight low-rise",
8:"Large low-rise",
9:"Sparsely built",
10:"Heavy industry",
11:"Dense trees",
12:"Scattered trees",
13:"Bush or scrub",
14:"Low plants",
15:"Bare rock or paved",
16:"Bare soil or sand",
17:"Water"}
df["name"] = df["LCZ"].map(lcz_names)
```

```python
df = df.sort_values("LCZ")
df
```
